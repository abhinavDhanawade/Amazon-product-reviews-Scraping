# Amazon Product Reviews Scraping

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
lnk = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page = requests.get(lnk)
data2 = page.text
soup= BeautifulSoup(data2,'lxml')

ama_review=soup.find_all('div',class_="a-section review aok-relative")

all_review={}
count = 0

for reviewed in ama_review:
    names = reviewed.span.text
    title = reviewed.find('a',class_='review-title-content').text
    rating = reviewed.find('i',class_='review-rating').text
    review = reviewed.find("span",{"data-hook":"review-body"}).text
    count+=1
    all_review[count]=[names,title,rating,review]

review_df= pd.DataFrame.from_dict(all_review,orient="index",columns=['names','title','rating','review'])

review_df
```

```python
review_df.to_csv('B:/matches_df.csv')
```
